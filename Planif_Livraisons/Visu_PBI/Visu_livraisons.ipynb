```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier
df = pd.read_excel('Planif livraisons.xlsx')

# Convertir la colonne "Date expédition" en datetime
df['Date expédition'] = pd.to_datetime(df['Date expédition'])

# Extraire l'année et le mois
df['Année'] = df['Date expédition'].dt.year
df['Mois'] = df['Date expédition'].dt.month

# Filtrer les données pour ne conserver que les années 2022, 2023, 2024
df_filtered = df[df['Année'].isin([2022, 2023, 2024])]

# Grouper par année et mois, et sommer la quantité livrée
quantite_livree_par_annee_mois = df_filtered.groupby(['Mois', 'Année'])['Qté livrée'].sum().unstack(fill_value=0)

# Tracer l'histogramme groupé
quantite_livree_par_annee_mois.plot(kind='bar', width=0.8, figsize=(12, 6), stacked=False)

# Ajouter des titres et labels
plt.title('Quantité livrée par mois et année')
plt.xlabel('Mois')
plt.ylabel('Quantité livrée')

# Personnaliser les étiquettes de l'axe des x pour afficher les mois
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=0)

# Ajouter une légende pour les années
plt.legend(title="Année", bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajouter une grille
plt.grid(True)

# Afficher le graphique
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier
df = pd.read_excel('Planif livraisons.xlsx')

# Grouper par désignation d'article et sommer les quantités livrées
quantite_livree_par_article = df.groupby('Désignation article')['Qté livrée'].sum()

# Si trop de catégories, regrouper les petites catégories sous 'Autres'
seuil = quantite_livree_par_article.sum() * 0.05  # seuil de 5% de la quantité totale
# Séparer les articles qui dépassent le seuil
quantite_livree_par_article = quantite_livree_par_article[quantite_livree_par_article >= seuil]

# Ajouter un élément "Autres" pour les petites catégories
autres_quantite = pd.Series(
    quantite_livree_par_article[quantite_livree_par_article < seuil].sum(), 
    index=['Autres']
)

# Utiliser pd.concat() pour combiner les "Autres" avec le reste
quantite_livree_par_article = pd.concat([quantite_livree_par_article, autres_quantite])

# Générer une palette de couleurs (par exemple, ici on utilise une palette de couleurs distinctes)
colors = plt.cm.Paired(range(len(quantite_livree_par_article)))

# Créer le graphique en secteur
plt.figure(figsize=(8, 8))
plt.pie(quantite_livree_par_article, labels=quantite_livree_par_article.index, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops={'edgecolor': 'black'})

# Ajouter un titre
plt.title('Quantité livrée par désignation d\'article')

# Ajouter une légende avec les couleurs correspondant à chaque article, en dehors du graphique
plt.legend(quantite_livree_par_article.index, title="Articles", loc='center left', bbox_to_anchor=(1, 0, 0.5, 1), fontsize=10)

# Afficher le graphique
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Charger le fichier
df = pd.read_excel('Planif livraisons.xlsx')

# Convertir la colonne "Date expédition" en datetime
df['Date expédition'] = pd.to_datetime(df['Date expédition'])

# Extraire l'année
df['Année'] = df['Date expédition'].dt.year

# Grouper par année et sommer les quantités commandées et livrées
quantite_totale_par_annee = df.groupby('Année').agg({
    'Qté cdée': 'sum',    # Quantité commandée
    'Qté livrée': 'sum'   # Quantité livrée
})

# Palette de couleurs distinctes pour chaque année et catégorie
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6']

# Créer un graphique en anneau pour les quantités commandées et livrées sur le même anneau
fig, ax = plt.subplots(figsize=(8, 8))

# Tracer les segments pour chaque année (Quantité commandée et livrée)
labels = []
sizes = []
colors_for_pie = []

# On construit les segments en ajoutant les quantités commandées et livrées pour chaque année
for year in quantite_totale_par_annee.index:
    commande = quantite_totale_par_annee.loc[year, 'Qté cdée']
    livree = quantite_totale_par_annee.loc[year, 'Qté livrée']
    
    sizes.append(commande)   # Quantité commandée
    sizes.append(livree)     # Quantité livrée
    labels.append(f"{year} Commandée")  # Label pour la quantité commandée
    labels.append(f"{year} Livrée")    # Label pour la quantité livrée
    colors_for_pie.append(colors[2 * (year - 2022)])  # Couleur pour la commande
    colors_for_pie.append(colors[2 * (year - 2022) + 1])  # Couleur pour la livraison

# Tracer le graphique en anneau
wedges, texts, autotexts = ax.pie(sizes, labels=labels, autopct='%1.2f%%', startangle=90,
                                  colors=colors_for_pie, wedgeprops={'width': 0.4, 'edgecolor': 'black'})

# Ajouter un titre principal
ax.set_title('Quantités commandées et livrées par année')

# Ajouter la légende
handles, labels = ax.get_legend_handles_labels()
# Personnaliser la légende pour qu'elle soit plus lisible
ax.legend(handles=handles, labels=labels, title="Année et Catégorie", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Ajuster l'affichage pour éviter les chevauchements
plt.tight_layout()

# Afficher le graphique
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier Excel
df = pd.read_excel('production.xlsx')

# Vérifier les premières lignes du fichier pour voir les types de données
print(df.head())

# Nettoyer la colonne 'Qté' pour garantir qu'elle contient uniquement des valeurs numériques
df['Qté'] = pd.to_numeric(df['Qté'], errors='coerce')

# Vérifier les lignes où la conversion a échoué (si des NaN apparaissent)
print(df[df['Qté'].isna()])

# Si des valeurs NaN apparaissent, vous pouvez les traiter :
df['Qté'] = df['Qté'].fillna(0)

# Grouper par client et sommer les quantités (Qté) sur tous les ans
quantites_par_client = df.groupby('Client')['Qté'].sum().reset_index()

# Trier les données par quantités pour avoir les clients les plus importants en premier
quantites_par_client = quantites_par_client.sort_values(by='Qté', ascending=False)

# Créer un graphique à barres pour la quantité totale par client
plt.figure(figsize=(12, 6))  # Taille du graphique
plt.bar(quantites_par_client['Client'], quantites_par_client['Qté'], color='lightcoral', edgecolor='black')

# Ajouter un titre et des labels
plt.title('Quantité Totale Commandée par Client (Tous les Ans)', fontsize=16)
plt.xlabel('Client', fontsize=12)
plt.ylabel('Quantité Totale', fontsize=12)

# Ajouter une grille sur l'axe des ordonnées
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Rotation des labels des clients pour éviter les chevauchements
plt.xticks(rotation=45, ha='right', fontsize=12)

# Ajuster l'affichage pour éviter les chevauchements
plt.tight_layout()

# Afficher le graphique
plt.show()
```
